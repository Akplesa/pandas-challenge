```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "../Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

```python
#PLAYER COUNT
#count total players using SN column and print into dataframe
number_players = purchase_data["SN"].nunique()
Total_players= pd.DataFrame({"Total_players": [number_players] })
Total_players
```

```python
#PURCHASING ANALYSIS
#Obtain number of unique items, average price, etc using nunique,count,sum, and mean functions and display in dataframe.
Item_ID = purchase_data["Item ID"].nunique()
Item_ID_Total= purchase_data["Item ID"].count()
Total_Revenue= purchase_data["Price"].sum()
Average_Price = purchase_data["Price"].mean()

Purchasing_analysis= pd.DataFrame({"Unique Items": [Item_ID],"Item ID Total ": [Item_ID_Total], 
                                   "Total Revenue": [Total_Revenue], "Average Price": [Average_Price]})
#Format so it looks pretty<3
Purchasing_analysis["Average Price"]=Purchasing_analysis["Average Price"].astype(float).map(
"${:,.2f}".format)
Purchasing_analysis["Total Revenue"]=Purchasing_analysis["Total Revenue"].astype(float).map(
"${:,.2f}".format)


Purchasing_analysis

```

```python
# Create Gender Demographics Percentage
Gender_percentage= purchase_data['Gender'].value_counts(normalize=True) * 100
#Create Gender Demographics Count
Gender_count= purchase_data['Gender'].value_counts()
Gender_Demo= pd.DataFrame({"Gender Percentage":Gender_percentage, "Gender count":Gender_count})

#Format Gender% to two decimal places
Gender_Demo["Gender Percentage"]=Gender_Demo["Gender Percentage"].astype(float).map(
"%{:,.2f}".format)

Gender_Demo
```

```python
#PURCHASING ANALYSIS (Gender)
#describe(by gender) ;) using groupby and count/mean functions and then display results in DF

Gender_Group= purchase_data.groupby("Gender")
Purchase_count= Gender_Group["Purchase ID"].count()
Purchase_Price_Average= Gender_Group["Price"].mean()
Purchase_Total= Gender_Group["Item ID"].count()
Purchasing_analysis_gender= pd.DataFrame({"Purchase Count": Purchase_count,
"Average Purchase Price": Purchase_Price_Average, "Purchase Total":Purchase_Total
                                         })
#Format so it looks pretty
#Convert the avg purchase price column to float and adjust values to 2 decimal places with $ sign. 
Purchasing_analysis_gender["Average Purchase Price"]=Purchasing_analysis_gender["Average Purchase Price"].astype(float).map(
"${:,.2f}".format)


Purchasing_analysis_gender
```

```python
#AGE DEMOGRAPHICS

#Before creating bins for ages first find the youngest, middle, and oldest ages.
Ages= purchase_data["Age"].describe() 
Ages

#Establish bins for ages
Age_Demographics= pd.DataFrame(purchase_data)
Age_range = [0, 12.9,19.9, 24.9, 39.9, 45 ]

#Catogerize players by "generation" using age bins
Group_generations= ["Kids", "Teenagers", "Gen Z", "Millennials","Gen X"]
Age_Demographics["Age Demographics"]= pd.cut(Age_Demographics["Age"], Age_range,labels= Group_generations, include_lowest= True)



Age_Demographics= Age_Demographics.groupby("Age Demographics")

Total= Age_Demographics["SN"].nunique().sum()

Total_count= Age_Demographics["SN"].count()
#ungrouped data for deno 
Age_demographic_percentage = (Age_Demographics["Age Demographics"].value_counts()/Total)*100

Age_demographic_percentage = Age_demographic_percentage.droplevel(0)




Purchasing_age_demographic= pd.DataFrame({"Total count": Total_count,

                                          "Age Demographic Percentage":Age_demographic_percentage   
                                         })
# Format clean up
#Convert the avg age demographic percentage column to float and adjust values to 2 decimal places with $ sign. 
Purchasing_age_demographic[ "Age Demographic Percentage"]=Purchasing_age_demographic[ "Age Demographic Percentage"].astype(float).map(
"${:,.2f}".format)

Purchasing_age_demographic
```

```python
#PURCHASING ANALYSIS BY AGE
#Before creating bins for ages first find the youngest, middle, and oldest ages.
Ages= purchase_data["Age"].describe() 
Ages

#Establish bins for ages
Age_Demographics= pd.DataFrame(purchase_data)
Age_range = [0, 12.9,19.9, 24.9, 39.9, 45 ]

#Catogerize players by "generation" using age bins
Group_generations= ["Kids", "Teenagers", "Gen Z", "Millennials","Gen X"]
By_Age_Group= Age_Demographics["Age Demographics"]= pd.cut(Age_Demographics["Age"], Age_range,labels= Group_generations, include_lowest= True)


# Group calculations by age demographic
Age_Group= Age_Demographics.groupby("Age Demographics")

Purchase_count_Age= Age_Group["Purchase ID"].count()
Purchase_Price_Average_Age= Age_Group["Price"].mean()
Purchase_Total_Age= Age_Group["Item ID"].count()
Average_Total_Per_Person= purchase_data["Price"].mean()

Purchasing_analysis_age= pd.DataFrame({"Purchase Count": Purchase_count_Age,
"Avg Purchase Price": Purchase_Price_Average_Age, "Purchase Total":Purchase_Total_Age, "Average Purchase Per Person":Average_Total_Per_Person
                                          
                                         })
#Cleaner Formating  
Purchasing_analysis_age["Avg Purchase Price"]= Purchasing_analysis_age["Avg Purchase Price"].astype(float).map(
"${:,.2f}".format)
Purchasing_analysis_age["Average Purchase Per Person"]= Purchasing_analysis_age["Average Purchase Per Person"].astype(float).map(
"${:,.2f}".format)
Purchasing_analysis_age
```

```python
#TOP SPENDERS
#Group data by SN

Spenders_SN= purchase_data.groupby("SN")
#Assign variables and run basic calculations
Purchase_price= Spenders_SN["Item ID"].nunique()
Item_Price= Spenders_SN["Price"].mean()
Total_value= Purchase_price*Item_Price
#Create dataframe and  sort total purchase value from highest to lowest price.
Top_Spenders= pd.DataFrame({"Purchase Count":Purchase_price, "Item Price":Item_Price, "Total Purchase Value":Total_value})
Spenders_df= Top_Spenders.sort_values("Total Purchase Value", ascending= False)

#Make Dataframe look pretty 
Spenders_df["Total Purchase Value"]= Spenders_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
Spenders_df["Item Price"]= Spenders_df["Item Price"].astype(float).map("${:,.2f}".format)

Spenders_df
```

```python
#MOST POPULAR ITEMS
#Locate columns 

Popular= purchase_data.set_index("Item ID", "Item Name","Price")
# Popular.head()

D_Name = purchase_data.groupby(["Item ID", "Item Name"])
P_count= D_Name["Item ID"].count()
Item_P=  D_Name["Price"].mean()
Total_Purchase_value= P_count*Item_P
Most_pop_items= pd.DataFrame({"Purchase Count":P_count, "Item Price":Item_P,"Total Purchase Value":Total_Purchase_value})
Items_df= Most_pop_items.sort_values("Purchase Count", ascending= False)

#Clean Data
Items_df["Total Purchase Value"]= Items_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
Items_df["Item Price"]= Items_df["Item Price"].astype(float).map("${:,.2f}".format)



Items_df
```

```python
#MOST PROFITABLE ITEMS
#Locate columns 

Popular= purchase_data.set_index("Item ID", "Item Name","Price")
# Popular.head()

D_Name = purchase_data.groupby(["Item ID", "Item Name"])
P_count= D_Name["Item ID"].count()
Item_P=  D_Name["Price"].mean()
Total_Purchase_value= P_count*Item_P
Most_pop_items= pd.DataFrame({"Purchase Count":P_count, "Item Price":Item_P,"Total Purchase Value":Total_Purchase_value})
Items_df= Most_pop_items.sort_values("Total Purchase Value", ascending= False)

#Clean Data
Items_df["Total Purchase Value"]= Items_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
Items_df["Item Price"]= Items_df["Item Price"].astype(float).map("${:,.2f}".format)



Items_df
```
